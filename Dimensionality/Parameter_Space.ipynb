# Parameter Space

Not all my parameters are free, many should be related. This notebook is used to investigate the parameters of my grid to see which ones need to be tied together to get a reasonable set.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from glob import glob
import random
```

# Column Densities
Column densities of the molecules and the overall gas are likely to be tied to each other in some manner. The av will also be tied to the other values.

```python
files=glob("../Benchmarking/grid_inputs/*")
df=pd.DataFrame()
for file in files:
    df=df.append(pd.read_csv(file),sort=False)
```

```python
fig,ax=plt.subplots()
sns.scatterplot(data=df,x="total_col_dens",y="total_h2_col",ax=ax)
sns.scatterplot(data=df,x="total_col_dens",y="total_c_col",ax=ax)
ax.plot(df["total_col_dens"],df["total_col_dens"])
ax.set(xscale="log",yscale="log",xlim=(1e9,1e25),ylim=(1e12,1e25))
ax.axvline(1e22)
ax.axvline(1e23)
ax.axhline(5e16)
```

Seems reasonable to generate a random column density and then use any value up to that column density for the species' column densities. We obviously work out $A_v$ from the total column density
